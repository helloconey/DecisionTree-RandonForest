## k—means is an unsupervise learning.
Goal: seperate to useful clusters 

Classic algoritms: K-means, DBSCAN, hierarchical clustering

Classification: decision tree, Bayes, logistic

Classification:       
- cluster.SffinityPropagation 
- cluster.Birch
- cluster.KMeans
- cluster.DBSCAN
- cluster.MiniBatchKMeans
- cluster.MeanShift
- cluster.SpectralClustering
- cluster.ward_tree
- cluster.mean_shift
- cluster.estimater_bandwidth

Logic:
1. randomly pick k samples as initial center
2. iteration.
3. put each sample and locate them to nearest initial center, will get k clusters
4. for each cluster, calculate the mean for all samples that allocated to this cluster. Then the mean become the new center.
5. When the location stop changed, stop iteration, we finished clustering.

- Euclidean distance： d(x, μ) = sqrt(sum(xi - μi)^2)
- Manhattan distance： d(x, μ) = sum(|xi - μi|)

if we use Euclidean distance, for each cluster, the distance of sum of square for every samples : cluster sum of square(CSS) = sum(sum(xi - μi)^2

- total cluster sum of square: sum(CSS)

```python
# self-create the dataset
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 4, random_state = 1)
                #create how many samples, how many features, how many centers

fig, ax1 = plt.subplots(1) #how many subplot? here is 1
ax1.scatter(X[:,0], X[:, 1], marker = 'o', s = 8) #marker=shape, s = size of shape
#scatter plot, extract 0 column as x, 1 column as y

plt.show()
```

```python
#KMeans does not need to build model and predict, we only need to fit
#KMeans has attribute predict and fit_predict (learn X and predict X)

pre = cluster.fit_predict(X)
pre == y_pred # so fit_predict and labels_ same
```

## *Important: to evaluate the performace of KMeans, we evaluate the outer/inner differencies of clusters to see performance.

##### inertia is used to measure the difference index within the cluster through distnace，close to 0 is better... 

**1.Contour coefficient** (when true label is unknown): range from -1 to 1. The closer to 1, the closer the sample is to the sample in the cluster where it belongs, and the samples in other clusters do not want to die. When the sample is similar to the out-cluster sample, the contour coefficient is negative. If it is 0, it means that the sample similarity between the two clusters is the same, and the two clusters should be the same cluster.

- s = (b-a)/ max(a,b)
- s = {1-a/b, if a<b;  0, if a=b;  b/a-1, if a>b }

- **in sklearn, we use metrics.silhouette_score to calculate contour coef**

## Handed Practice: using Contour Coef to choose best n_clusters:

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt 
import matplotlib.cm as cm 
import numpy as np
```

```python
n_clusters = 4

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(18, 7)
ax1.set_xlim([-0.1, 1])
ax1.set_ylim([0, X.shape[0] + (n_clusters + 1) * 10])
clusterer = KMeans(n_clusters=n_clusters, random_state=1).fit(X) 
cluster_labels = clusterer.labels_

silhouette_avg = silhouette_score(X, cluster_labels)
print("For n_clusters =", n_clusters,
      "The average silhouette_score is :", silhouette_avg)
sample_silhouette_values = silhouette_samples(X, cluster_labels)

y_lower = 10
for i in range(n_clusters):
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i] 
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    color = cm.nipy_spectral(float(i)/n_clusters) 
    ax1.fill_betweenx(np.arange(y_lower, y_upper)
    ,ith_cluster_silhouette_values ,facecolor=color
    ,alpha=0.7
    ) 
    ax1.text(-0.05
    , y_lower + 0.5 * size_cluster_i
             , str(i))
    y_lower = y_upper + 10

    
    
ax1.set_title("The silhouette plot for the various clusters.") 
ax1.set_xlabel("The silhouette coefficient values") 
ax1.set_ylabel("Cluster label")

ax1.axvline(x=silhouette_avg, color="red", linestyle="--") 
ax1.set_yticks([])
ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])


colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)


ax2.scatter(X[:, 0], X[:, 1] ,marker='o'
            ,s=8
            ,c=colors )

centers = clusterer.cluster_centers_
# Draw white circles at cluster centers 
ax2.scatter(centers[:, 0], centers[:, 1], marker='x',
            c="red", alpha=1, s=200)

ax2.set_title("The visualization of the clustered data.") 
ax2.set_xlabel("Feature space for the 1st feature") 
ax2.set_ylabel("Feature space for the 2nd feature")


plt.suptitle(("Silhouette analysis for KMeans clustering on sample data " 
              "with n_clusters = %d" % n_clusters),fontsize=14, fontweight='bold')
plt.show()
```
